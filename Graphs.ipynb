```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
x=np.linspace(0,10,100)
y=np.sin(x)
plt.plot(x,y,label='sin(x)',color='blue',linestyle='--')
plt.title('Line plot')
plt.xlabel("X=axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
plt.scatter(x,y,color='cyan',alpha=0.7)
plt.title("Scatter plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
categories=['A','B','C','D']
values=[10,20,30,40]
plt.bar(categories,values,color='blue')
plt.title("Bar plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
data=np.random.rand(1000)
plt.hist(data,bins=30,color='purple',alpha=0.8)
plt.title("Histogram")
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.show()
```

```python
labels=['Python','Java','C++','Ruby']
sizes=[50,70,80,10]
plt.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=140)
plt.title("Pie chart")
plt.show()
```

```python
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1,label='sin(x)',color='red',linewidth=2,marker='o')
plt.plot(x,y2,label='cos(x)',color='blue',linestyle='dotted')
plt.title("Line plot")
plt.xlabel("X=Axis")
plt.ylabel("Y-Axis")
plt.legend()
plt.grid(True,linestyle='--',color='gray',alpha=0.6)
plt.show()
```
