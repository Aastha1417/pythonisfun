```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Random scores between 0 and 100
num_students = 100
num_subjects = 5
scores = np.random.randint(0, 101, size=(num_students, num_subjects))
```

```python
# Displaying the generated scores
print("Scores of 100 Students in 5 Subjects:")
print(scores)
```

```python
# Addition
addition_result = scores + scores
print("\nAddition Result:")
print(addition_result)
```

```python
# Subtraction
subtraction_result = scores - scores
print("\nSubtraction Result:")
print(subtraction_result)
```

```python
# Multiplication
multiplication_result = scores * scores
print("\nMultiplication Result:")
print(multiplication_result)
```

```python
# Division
division_result = scores / scores
print("\nDivision Result:")
print(division_result)
```

```python
# Horizontal Stacking (hstack)
new_subject_scores = np.random.randint(0, 101, size=(num_students, 1))
hstack_result = np.hstack((scores, new_subject_scores))
print("\nHorizontal Stacking Result:")
print(hstack_result)
```

```python
# Mean
mean_scores = np.mean(scores, axis=0)
print("\nMean Scores for Each Subject:")
print(mean_scores)
```

```python
# Median
median_scores = np.median(scores, axis=0)
print("\nMedian Scores for Each Subject:")
print(median_scores)
```

```python
# Standard Deviation
std_dev_scores = np.std(scores, axis=0)
print("\nStandard Deviation of Scores for Each Subject:")
print(std_dev_scores)

```

```python
# Finding the maximum and minimum scores for each subject
max_scores = np.max(scores, axis=0)
min_scores = np.min(scores, axis=0)
print("\nMaximum Scores for Each Subject:")
print(max_scores)
print("\nMinimum Scores for Each Subject:")
print(min_scores)
```

```python
# Reshaping the scores array (for demonstration)
reshaped_scores = scores.reshape(20, 25)  
print("\nReshaped Scores Array (20x25):")
print(reshaped_scores)
```

```python
# Slicing: Get scores of the first 10 students
sliced_scores = hstack_result[:10, :]  # First 10 students
print("\nSliced Scores of the First 10 Students:")
print(sliced_scores)
```

```python
# Visualizing the average scores
plt.figure(figsize=(10, 6))
plt.bar(range(1, num_subjects + 1), mean_scores, color='skyblue')
plt.xticks(range(1, num_subjects + 1), [f'Subject {i+1}' for i in range(num_subjects)])
plt.title('Average Scores for Each Subject')
plt.xlabel('Subjects')
plt.ylabel('Average Score')
plt.ylim(0, 100)
plt.grid(axis='y')
plt.show()
```

```python
plt.subplot(2, 2, 4)
for i in range(num_subjects):
    plt.scatter(range(num_students), hstack_result[:, i], alpha=0.5, label=f'Subject {i+1}')
plt.title('Scatter Plot of Scores for Each Student')
plt.xlabel('Student Index')
plt.ylabel('Scores')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

```python
plt.subplot(2, 2, 3)
plt.boxplot(hstack_result, labels=[f'Subject {i+1}' for i in range(num_subjects + 1)])
plt.title('Boxplot of Scores for Each Subject')
plt.ylabel('Scores')
plt.grid(axis='y')
```
